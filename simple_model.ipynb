# Inputs

```python
salary = 60_000
savings_rate = 0.25
investment_rate = 0.05
desired_cash = 1_500_00
```

# Determining Cash Saved

```python
annual_cash = salary * savings_rate
annual_cash
```

# Years to Retirement

```python
print(annual_cash)
```

Финансовые функции убраны из `numpy`, теперь они в отдельном модуле `numpy_financial`

```python
import numpy_financial as npf
```

```python
years_to_retirement = npf.nper(investment_rate, -annual_cash, 0, desired_cash)
```

```python
print(f"Martha has {years_to_retirement: .1f} years to retirement.")
```
